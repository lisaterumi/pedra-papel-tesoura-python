### Código completo:


````python
import random
import os
import re
os.system('cls' if os.name=='nt' else 'clear')

# imprime o nome inteiro da escolha
def choice(input):
    if (input == 'PA'):
        return 'PAPEL'
    elif (input == 'PE'):
        return 'PEDRA'
    elif (input == 'T'):
        return 'TESOURA'

while (1 < 2):
    print ("\n")
    print ("Pedra, Papel, Tesoura - Que comece o jogo!")
    userChoice = input("Escolha: [PE]dra, [PA]pel, ou [T]esoura: ")
    if not re.match("[PEpePApaTt]", userChoice):
        print ("Por favor, informe uma opção abaixo:")
        print ("[PE]dra, [PA]pel, ou [T]esoura.")
        continue
    #Escolha do usuário
    print ("Sua escolha: " + choice(userChoice.upper()))
    choices = ['PE', 'PA', 'T']
    opponenetChoice = random.choice(choices)
    print ("Eu escolho: " + choice(opponenetChoice.upper()))
    if opponenetChoice == str.upper(userChoice):
        print ("Empate! ")
    #if opponenetChoice == str("R") and str.upper(userChoice) == "P"
    elif opponenetChoice == 'PE' and userChoice.upper() == 'T':      
        print ("Pedra ganha de tesoura, eu ganhei!")
        continue
    elif opponenetChoice == 'T' and userChoice.upper() == 'PA':      
        print ("Tesoura ganha de pedra, eu ganhei!")
        continue
    elif opponenetChoice == 'PA' and userChoice.upper() == 'PE':      
        print ("Papel ganha de pedra, eu ganhei!")
        continue
    else:       
        print ("Você venceu!")
````
